# 前言
对pandas.Series.plot()方法进行详细解析

```python
# 引入包
import numpy as np
import pandas as pd
# 行内显示图像
%matplotlib inline
```

```python
# 简单查看下plot方法包含的信息
help(pd.Series.plot)
```

```python
# 构建一个时间序列，值随机，命名为randval -- random value
series = pd.Series(np.random.randn(10),index= pd.date_range('2019-3-1',periods=10,freq='D'),name = 'randval')
series   
```

```python
# 默认就是使用折线图
# Series.plot(kind='line') 和 Series.plot.line() 是等价的，其他的图形也是一样
series.plot()
(series+1).plot(kind='line')
(series+2).plot.line()
```

```python
# 图例
series.plot(legend=True)
```

```python
# 图例中列的名称用 label 参数来控制
series.plot(legend=True,label='testvalue')
```

```python
# 横轴坐标默认使用index，如果设置为False，则横坐标为序号
series.plot(legend=True,use_index=True)
```

```python
# 网格，标题，图像大小(需要特别注意单位是英寸)，字体大小，横轴刻度旋转
series.plot(legend=True,use_index=True,grid=True,title='timesequence',figsize=(6,4),fontsize=14,rot=30)
```

```python
# 折线图的线型
series.plot(style='-',legend=True,use_index=True,title='timesequence',fontsize=14,label='line')
(series-1).plot(style='-.',legend=True,use_index=True,title='timesequence',fontsize=14,label='dot_line')
(series-2).plot(style=':',legend=True,use_index=True,title='timesequence',fontsize=14,label='dot')
```

```python
# 其他参数**kwads
# 因为还可以接受matplotlib的参数，这里指定各折线的颜色
series.plot(style='-',legend=True,use_index=True,title='timesequence',fontsize=14,label='line',color='red')
(series-1).plot(style='-.',legend=True,use_index=True,title='timesequence',fontsize=14,label='dot_line',color='blue')
(series-2).plot(style=':',legend=True,use_index=True,title='timesequence',fontsize=14,label='dot',color='green')
```

```python
# 设置纵轴坐标刻度线 yticks， 横轴同理
series.plot(style='-',legend=True,use_index=True,title='timesequence',label='line',color='red',
    xticks=[pd.to_datetime('2019-3-3'),pd.to_datetime('2019-3-7')],yticks=[-2,-1.5,-1,-0.5,0,0.5,1])
(series-1).plot(style='-.',legend=True,use_index=True,title='timesequence',label='dot_line',color='blue')
(series-2).plot(style=':',legend=True,use_index=True,title='timesequence',label='dot',color='green')
```

```python
# 表格，表格显示将横坐标值表示出来
series.plot(table=True)
```

```python
# 柱形图
series.plot(kind='bar')
```

```python
# 柱形图带误差线yerr， 横向误差线xerr同理
series.plot(kind='bar',yerr=True)
```

```python
# 横向柱形图
series.plot(kind='barh')
```

```python
# 直方图
series.plot.hist()
```

```python
# 箱线图
series.plot.box()
```

```python
# 密度图
series.plot.kde()
(series+1).plot.density()
```

```python
# 面积图 
# 注意点：需要所有值都为正 或都为负
(series+3).plot.area()
```

```python
# 饼图
# 注意点：需要所有值为正
(series+3).plot.pie(figsize=(4,4))
```
